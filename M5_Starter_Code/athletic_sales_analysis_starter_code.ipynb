```python
# Import Libraries and Dependencies
import pandas as pd
from pathlib import Path
from datetime import datetime
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletics_sales_2020_path = Path('Resources/athletic_sales_2020.csv')
athletics_sales_2021_path = Path('Resources/athletic_sales_2021.csv')

athletic_sales_2020_df  = pd.read_csv(athletics_sales_2020_path)
athletic_sales_2021_df  = pd.read_csv(athletics_sales_2021_path)
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
athletic_sales_2020_2021_df = pd.concat([athletic_sales_2020_df, athletic_sales_2021_df],axis=0,ignore_index = True)
athletic_sales_2020_2021_df.head()
athletic_sales_2020_2021_df.tail()
```

```python
# Check if any values are null.
athletic_sales_2020_2021_df.isna()
athletic_sales_2020_2021_df.dtypes
athletic_sales_2020_2021_df.info()
```

```python
# Check the data type of each column
athletic_sales_2020_2021_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
def isNaN(num):
    return num != num

v = []
for i in athletic_sales_2020_2021_df['invoice_date']:
    if(isNaN(i)):
        i='9/9/09'
    v.append(datetime.strptime(i,'%x') )
athletic_sales_2020_2021_df['invoice_date'] = v
athletic_sales_2020_2021_df.dtypes



```

```python
# Confirm that the "invoice_date" data type has been changed.
athletic_sales_2020_2021_df.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
grouped_df = athletic_sales_2020_2021_df.groupby(['region','state','city']).agg({'units_sold':'sum'})
grouped_df = grouped_df.rename(columns={'units_sold':'Total_Products_Sold'})

# Show the top 5 results.
grouped_df.sort_values('Total_Products_Sold', ascending = False).head(5)
                            
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
pivot_table_df = athletic_sales_2020_2021_df.pivot_table(values = 'units_sold', index   = ['region', 'state', 'city'], aggfunc='sum')

# Rename the "units_sold" column to "Total_Products_Sold"
pivot_table_df = pivot_table_df.rename(columns={'units_sold':'Total_Products_Sold'})

# Show the top 5 results.
pivot_table_df.sort_values('Total_Products_Sold', ascending = False).head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total Sales"
grouped_df = athletic_sales_2020_2021_df.groupby(['region','state','city']).agg({'units_sold':'sum','total_sales':'sum'})
grouped_df = grouped_df.rename(columns={'units_sold':'Total_Products_Sold','total_sales':'Total Sales' })

# Show the top 5 results.
grouped_df.sort_values('Total Sales', ascending = False).head(5)

```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
pivot_table_df = athletic_sales_2020_2021_df.pivot_table(values = ['total_sales','units_sold',], index   = ['region', 'state', 'city'], aggfunc='sum')
pivot_table_df = pivot_table_df.rename(columns={'units_sold':'Total_Products_Sold', 'total_sales':'Total Sales' })

# Show the top 5 results.
pivot_table_df.sort_values('Total Sales', ascending = False).head(5)

```

### 4. Determine which Retailer had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
# Rename the "total_sales" column to "Total Sales"
grouped_df = athletic_sales_2020_2021_df.groupby(['retailer','region','state','city']).agg({'units_sold':'sum','total_sales':'sum'})
grouped_df = grouped_df.rename(columns={'units_sold':'Total_Products_Sold','total_sales':'Total Sales' })

# Show the top 5 results.
grouped_df.sort_values('Total Sales', ascending = False).head(5)

```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each retailer, region, state, and city.
pivot_table_df = athletic_sales_2020_2021_df.pivot_table(values = ['total_sales','units_sold',], index   = ['retailer','region', 'state', 'city'], aggfunc='sum')

# Rename the "units_sold" column to "Total_Products_Sold"
pivot_table_df = pivot_table_df.rename(columns={'units_sold':'Total_Products_Sold', 'total_sales':'Total Sales' })

# Show the top 5 results.
pivot_table_df.sort_values('Total Sales', ascending = False).head(5)
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_athletic_footwear_df = athletic_sales_2020_2021_df.loc[athletic_sales_2020_2021_df['product']=="Women's Athletic Footwear"]
womens_athletic_footwear_df
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
grouped_df = womens_athletic_footwear_df.groupby(['retailer','region','state','city']).agg({'units_sold':'sum'})
grouped_df = grouped_df.rename(columns={'units_sold':'Womens_Footwear_Units_Sold'})

# Show the top 5 results.
grouped_df.sort_values('Womens_Footwear_Units_Sold', ascending = False).head(5)

```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
pivot_table_df = womens_athletic_footwear_df.pivot_table(values = ['units_sold',], index   = ['retailer','region', 'state', 'city'], aggfunc='sum')

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
pivot_table_df = pivot_table_df.rename(columns={'units_sold':'Womens_Footwear_Units_Sold' })

# Show the top 5 results.
pivot_table_df.sort_values('Womens_Footwear_Units_Sold', ascending = False).head(5)

```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
pivot_table_df = womens_athletic_footwear_df.pivot_table(values = ['total_sales',], index = ['invoice_date'], aggfunc='sum')


# Optional: Rename the "total_sales" column to "Total Sales"
pivot_table_df = pivot_table_df.rename(columns={'total_sales':'Total Sales'})

# Show the table.
pivot_table_df.sort_values('Total Sales', ascending = False).head(5)
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
pivot_table_df = pivot_table_df.resample('D').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
pivot_table_df.sort_values('Total Sales', ascending = False).head(10)
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
pivot_table_df = pivot_table_df.resample('W').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
pivot_table_df.sort_values('Total Sales', ascending = False).head(10)
```

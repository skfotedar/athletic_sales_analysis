```python
# Import Libraries and Dependencies
import pandas as pd
from pathlib import Path
from datetime import datetime
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletics_sales_2020_path = Path('Resources/athletic_sales_2020.csv')
athletics_sales_2021_path = Path('Resources/athletic_sales_2021.csv')

athletic_sales_2020_df  = pd.read_csv(athletics_sales_2020_path)
athletic_sales_2021_df  = pd.read_csv(athletics_sales_2021_path)
```

```python
# Display the 2020 sales DataFrame
athletic_sales_2020_df.head(20)
```

```python
# Display the 2021 sales DataFrame
athletic_sales_2021_df.head(20)
```

```python
# Check the 2021 sales data types.
athletic_sales_2021_df.dtypes 
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
athletic_sales_2020_2021_df = athletic_sales_2021_df.join(athletic_sales_2020_df, lsuffix='_2021', rsuffix = '_2020')
athletic_sales_2020_2021_df.reset_index()
athletic_sales_2020_2021_df.head()
```

```python
# Check if any values are null.
athletic_sales_2020_2021_df.isna()
athletic_sales_2020_2021_df.info()
```

```python
# Check the data type of each column
athletic_sales_2020_2021_df.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
def isNaN(num):
    return num != num

v = []
for i in athletic_sales_2020_2021_df.invoice_date_2020:
    if(isNaN(i)):
        i='1/1/01'
    v.append(datetime.strptime(i,'%x') )
athletic_sales_2020_2021_df['invoice_date_2020'] = v

v = []
for i in athletic_sales_2020_2021_df.invoice_date_2021:
    if(isNaN(i)):
        i='1/1/01'
    v.append(datetime.strptime(i,'%x') )
athletic_sales_2020_2021_df['invoice_date_2021'] = v
athletic_sales_2020_2021_df['invoice_date_2021']

```

```python
# Confirm that the "invoice_date" data type has been changed.
print(athletic_sales_2020_2021_df.dtypes['invoice_date_2021'])
print(athletic_sales_2020_2021_df.dtypes['invoice_date_2020'])
```
